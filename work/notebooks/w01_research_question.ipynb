```python
df = pd.read_csv('data/raw/content_refresh_anonymized.csv')

declining_visible = df[(df['trend_direction']=='down') & (df['impressions_90d']>=100)]
print('Declining pages with real demand (impressions_90d>=100):', len(declining_visible), '/', len(df))

stale_visible = df[(df['days_since_last_update']>=180) & (df['impressions_90d']>=500)]
print('Stale but visible pages:', len(stale_visible), '/', len(df))

decay_risk = df[(df['avg_position']>0) & (df['avg_position']<=10) & (df['content_age_days']>=180)]
print('Page-one decay risk pages:', len(decay_risk), '/', len(df))
```

```python
"""
On the 30,000-row starter sample:
- 13,152 pages are declining with real demand ('trend_direction ==
  'down'' and 'impressions_90d >= 100') — a large enough pool that ranking
  which one to review first is a genuine problem, not a rounding error.

- Only 17 pages are both stale (no update in 180+ days) and still
  getting meaningful traffic ('impressions_90d >= 500') — a much smaller,
  higher-confidence signal on its own, which tells me 'days_since_last_update'
  alone is too narrow a rule and supports combining signals rather than
  picking one threshold.

  - 7,076 pages sit in the top 10 positions but are old enough
  ('content_age_days >= 180') to be at decay risk — a meaningful chunk of
  "protect this" candidates, separate from the decline group.

"""
```
