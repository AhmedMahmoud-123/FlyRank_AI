## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

Using a target that's directly observed (measured independently of the model's features) is the best way to simulate the real-world decision the model needs to support. If that observed outcome isn't available — because it's delayed, expensive to collect, or simply not tracked — we have to fall back on a proxy, and hope it behaves similarly to the real target. The risk is that a proxy built from the same columns available as features stops being a genuine target and becomes something the model can just reverse-engineer rather than predict.
